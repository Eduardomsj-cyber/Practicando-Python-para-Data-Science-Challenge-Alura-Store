### Importantando librerias

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
```

# **1. Análisis de facturación**

```python
tienda.head()
```

# **Ventas por categoría**

En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.





```python
ventas_por_categoria_tienda = (
    tienda
    .groupby(["Lugar de Compra", "Categoría del Producto"])
    .size()
    .reset_index(name="Numero_de_ventas")
    .sort_values(["Lugar de Compra", "Numero_de_ventas"], ascending=[True, False])
)

ventas_por_categoria_tienda2 = (
    tienda2
    .groupby(["Lugar de Compra", "Categoría del Producto"])
    .size()
    .reset_index(name="Numero_de_ventas")
    .sort_values(["Lugar de Compra", "Numero_de_ventas"], ascending=[True, False])
)

ventas_por_categoria_tienda3 = (
    tienda3
    .groupby(["Lugar de Compra", "Categoría del Producto"])
    .size()
    .reset_index(name="Numero_de_ventas")
    .sort_values(["Lugar de Compra", "Numero_de_ventas"], ascending=[True, False])
)

ventas_por_categoria_tienda4 = (
    tienda4
    .groupby(["Lugar de Compra", "Categoría del Producto"])
    .size()
    .reset_index(name="Numero_de_ventas")
    .sort_values(["Lugar de Compra", "Numero_de_ventas"], ascending=[True, False])
)

ventas_por_categoria_tienda

ventas_por_categoria_tienda2

ventas_por_categoria_tienda3

ventas_por_categoria_tienda4

total = pd.concat([ventas_por_categoria_tienda, ventas_por_categoria_tienda2, ventas_por_categoria_tienda3, ventas_por_categoria_tienda4])
total
```

# **3. Calificación promedio de la tienda**

```python
Calificacón = tienda.groupby(['Categoría del Producto', 'Calificación']).size().reset_index(name="Numero_de_ventas").sort_values(["Numero_de_ventas"], ascending=[False]).head(10)
Calificacón
```

# **4. Productos más y menos vendidos**

```python
Mayor_Ventas= tienda.groupby(['Categoría del Producto', 'Calificación']).size().reset_index(name="Numero_de_ventas").sort_values(["Numero_de_ventas"], ascending=[False]).head(10)
Mayor_Ventas
```

```python
MENOR_Ventas= tienda.groupby(['Categoría del Producto', 'Calificación']).size().reset_index(name="Costo de envío").sort_values(["Costo de envío"], ascending=[False]).head(10)
MENOR_Ventas
```

# **5. Envío promedio por tienda**

```python
Calificacón.set_index('Categoría del Producto')['Numero_de_ventas'].plot(kind='pie', figsize=(10, 10), autopct='%1.1f%%', title='Número de Ventas por Categoría')
```

```python
Mayor_Ventas.plot(kind='bar', figsize=(10, 6),color='red')
```
